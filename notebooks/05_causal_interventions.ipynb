# 05 - Causal interventions (Phase 4)

Activation-steering effect at point E: Δ entropy / Δ margin vs alpha for the
emotion direction and the controls (opposite emotion, confounder, random
same-norm), plus ablation. Transient forward hooks; no weight change.

Run `python scripts/run_steering.py` first.

```python
import json
from pathlib import Path
ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
r = json.loads((ROOT / 'outputs/reports/steering_effects.json').read_text(encoding='utf-8'))
print('model:', r['model_id'], '| layer:', r['layer'], '| concept:', r['concept'])
print('alpha grid:', r['alpha_grid'])
```

```python
import matplotlib.pyplot as plt
tag = 'severe'
for cond, curve in r['inputs'][tag]['add'].items():
    plt.plot([c['alpha'] for c in curve], [c['delta_entropy'] for c in curve], marker='o', label=cond)
plt.axhline(0, color='gray', ls=':'); plt.axvline(0, color='gray', ls=':')
plt.xlabel('alpha (norm-scaled)'); plt.ylabel('Δ entropy at point E'); plt.title(f'{tag} input'); plt.legend(); plt.show()
print('ablation:', r['inputs'][tag]['ablation'])
```

```python
# how often does steering flip the top-1 decision token, by condition?
for tag in r['inputs']:
    print(tag)
    for cond, curve in r['inputs'][tag]['add'].items():
        flips = sum(c['top1_changed'] for c in curve)
        print(f'  {cond:12} top1 flips over alpha grid: {flips}/{len(curve)}')
```
