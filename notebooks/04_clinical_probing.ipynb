# 04 - Clinical probing at point E (Phase 3)

Emotion-like signal (z vs neutral) at the pre-decision point E vs symptom
severity, and emotion-vs-confounder distinguishability.

Run `python scripts/run_probing.py` first.

```python
import json
from pathlib import Path
ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
r = json.loads((ROOT / 'outputs/reports/clinical_probing.json').read_text(encoding='utf-8'))
print('model:', r['model_id'], '| layers used:', r['layer_of'])
print('\nGradient trends (Pearson step vs z):')
for g, tr in r['gradient_trends_pearson_step_vs_z'].items():
    print(g, {k: tr[k] for k in ['afraid_alarmed','anxious_nervous','general_negative_valence','safety_policy']})
```

```python
print('Distinguishability (emotion-z vs confounder-z):')
for k, v in r['distinguishability_emotion_vs_confounder'].items():
    if k.startswith('afraid') or k.startswith('anxious'):
        print(f'  {k:44} {v:+.3f}')
print('\nPersistence retained fraction:', r['persistence_retained_fraction'])
```

```python
# z vs severity step for a chosen gradient
import matplotlib.pyplot as plt
items = [v for v in r['per_item'].values() if v['group']=='gradient:nausea']
items = sorted(items, key=lambda x: x['step'])
steps = [x['step'] for x in items]
for c in ['afraid_alarmed','anxious_nervous','general_negative_valence','safety_policy']:
    plt.plot(steps, [x['z'][c] for x in items], marker='o', label=c)
plt.axhline(0, color='gray', ls=':'); plt.xlabel('severity step'); plt.ylabel('z vs neutral'); plt.legend(); plt.show()
```
